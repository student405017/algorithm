```python
from collections import deque

def fmt_stack(s):
    # list as stack: left=bottom, right=top
    return "[" + " ".join(map(str, s)) + "] (bottom->top)"

def fmt_queue(q):
    # deque as queue: left=front, right=back
    return "[" + " ".join(map(str, q)) + "] (front->back)"

def sort_with_stacks_trace_short(nums):
    """
    Stack sort (ascending) using one auxiliary stack.
    Print state once per insertion (short trace).
    Worst time: O(n^2), Space: O(n)
    """
    s = nums[:]   # input stack, top at end
    aux = []

    print("\n=== Stack Sort (short trace, ascending) ===")
    print("initial s  :", fmt_stack(s))
    print("initial aux:", fmt_stack(aux))

    step = 1
    while s:
        cur = s.pop()
        while aux and aux[-1] > cur:
            s.append(aux.pop())
        aux.append(cur)

        # print once per insertion completion
        print(f"after insert #{step} (cur={cur}): s={fmt_stack(s)} | aux={fmt_stack(aux)}")
        step += 1

    out = aux[::-1]  # ascending output
    print("final output:", out)
    print("時間複雜度: Worst O(n^2), Space O(n)")
    return out

def sort_with_queues_trace_short(nums):
    """
    Queue insertion-style sort (ascending) using one auxiliary queue.
    Print state once per insertion (short trace).
    Worst time: O(n^2), Space: O(n)
    """
    q = deque(nums)
    sorted_q = deque()

    print("\n=== Queue Sort (short trace, ascending) ===")
    print("initial q     :", fmt_queue(q))
    print("initial sorted:", fmt_queue(sorted_q))

    step = 1
    while q:
        x = q.popleft()

        inserted = False
        m = len(sorted_q)
        for _ in range(m):
            y = sorted_q.popleft()
            if (not inserted) and (y > x):
                sorted_q.append(x)
                inserted = True
            sorted_q.append(y)
        if not inserted:
            sorted_q.append(x)

        # print once per insertion completion
        print(f"after insert #{step} (x={x}): q={fmt_queue(q)} | sorted={fmt_queue(sorted_q)}")
        step += 1

    out = list(sorted_q)
    print("final output:", out)
    print("時間複雜度: Worst O(n^2), Space O(n)")
    return out

def main():
    n = int(input("請輸入 n (數字個數): ").strip())
    nums = list(map(int, input(f"請輸入 {n} 個整數(空白分隔): ").split()))
    if len(nums) != n:
        raise ValueError(f"你輸入了 {len(nums)} 個數字，但 n={n}")

    print("\n原始輸入:", nums)

    stack_sorted = sort_with_stacks_trace_short(nums)
    queue_sorted = sort_with_queues_trace_short(nums)

    print("\n=== Summary ===")
    print("Stack sorted:", stack_sorted)
    print("Queue sorted:", queue_sorted)

if __name__ == "__main__":
    main()
```
