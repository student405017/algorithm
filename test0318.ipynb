```python
# Stack 實作 (堆疊)
# LIFO: Last In, First Out

class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):       # 推入元素
        self.items.append(item)

    def pop(self):              # 彈出元素
        if not self.is_empty():
            return self.items.pop()
        return None

    def peek(self):             # 查看頂端元素
        if not self.is_empty():
            return self.items[-1]
        return None

    def is_empty(self):         # 檢查是否為空
        return len(self.items) == 0

    def size(self):
        return len(self.items)


s = Stack()
s.push(1)
s.push(2)
s.push(3)
print("Stack top (peek):", s.peek())
print("Pop:", s.pop())
print("Pop:", s.pop())
print("Stack size:", s.size())
print("時間複雜度: push O(1), pop O(1), peek O(1)")
```

```python
# Queue 實作 (佇列)
# FIFO: First In, First Out

from collections import deque

class Queue:
    def __init__(self):
        self.items = deque()

    def enqueue(self, item):    # 加入元素到隊尾
        self.items.append(item)

    def dequeue(self):          # 移除並回傳隊首元素
        if not self.is_empty():
            return self.items.popleft()
        return None

    def front(self):            # 查看隊首元素
        if not self.is_empty():
            return self.items[0]
        return None

    def is_empty(self):
        return len(self.items) == 0

    def size(self):
        return len(self.items)


q = Queue()
q.enqueue(1)
q.enqueue(2)
q.enqueue(3)
print("Queue front:", q.front())
print("Dequeue:", q.dequeue())
print("Dequeue:", q.dequeue())
print("Queue size:", q.size())
print("時間複雜度: enqueue O(1), dequeue O(1), front O(1)")
```

```python
# 用兩個 Stack 實作 Queue
# enqueue: O(1)
# dequeue: 均攤 O(1) (amortized O(1))

class QueueUsingStacks:
    def __init__(self):
        self.stack_in = []   # 負責 enqueue
        self.stack_out = []  # 負責 dequeue

    def enqueue(self, item):
        self.stack_in.append(item)

    def _transfer(self):
        # 當 stack_out 為空時，將 stack_in 的元素全部移到 stack_out
        if not self.stack_out:
            while self.stack_in:
                self.stack_out.append(self.stack_in.pop())

    def dequeue(self):
        self._transfer()
        if self.stack_out:
            return self.stack_out.pop()
        return None

    def front(self):
        self._transfer()
        if self.stack_out:
            return self.stack_out[-1]
        return None

    def is_empty(self):
        return not self.stack_in and not self.stack_out


q2 = QueueUsingStacks()
q2.enqueue("A")
q2.enqueue("B")
q2.enqueue("C")
print("Front:", q2.front())
print("Dequeue:", q2.dequeue())
q2.enqueue("D")
print("Dequeue:", q2.dequeue())
print("Dequeue:", q2.dequeue())
print("時間複雜度: enqueue O(1), dequeue 均攤 O(1)")
```

```python
# 用兩個 Queue 實作 Stack
# push: O(n)
# pop: O(1)

from collections import deque

class StackUsingQueues:
    def __init__(self):
        self.q1 = deque()  # 主要佇列
        self.q2 = deque()  # 輔助佇列

    def push(self, item):
        # 新元素加到 q2，再將 q1 全部移到 q2，最後交換 q1 與 q2
        self.q2.append(item)
        while self.q1:
            self.q2.append(self.q1.popleft())
        self.q1, self.q2 = self.q2, self.q1

    def pop(self):
        if self.q1:
            return self.q1.popleft()
        return None

    def peek(self):
        if self.q1:
            return self.q1[0]
        return None

    def is_empty(self):
        return len(self.q1) == 0


s2 = StackUsingQueues()
s2.push(10)
s2.push(20)
s2.push(30)
print("Stack top (peek):", s2.peek())
print("Pop:", s2.pop())
print("Pop:", s2.pop())
print("時間複雜度: push O(n), pop O(1), peek O(1)")
```
