```python
def mergSort(arr):
    if len(arr)<=1:
        return arr
    mid=len(arr)//2
    left=mergSort(arr[:mid])
    right=mergSort(arr[mid:])
    
    result=[]
    i=j=0
    while i<len(left) and j <len(right):
        if left[i]<right[j]:
            result.append(left[i])
            i+=1
        else:
            result.append(right[j])
            j+=1
    result +=left[i:]
    result +=right[j:]
    
    return result
    
arr=[38,21,33,5,31,12]
print(mergSort(arr))
print("時間複雜度 O(n log n)")
```

```python
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1

        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1

        arr[j + 1] = key
    return arr


arr = [38, 27, 43, 3, 9, 82, 10]
sorted_arr = insertion_sort(arr)

print("Sorted array:", sorted_arr)
print("Time Complexity of Insertion Sort: O(n²)")

```

```python
def bubble_sort(arr):
    n = len(arr)

    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]

    return arr


arr = [38, 27, 43, 3, 9, 82, 10]
sorted_arr = bubble_sort(arr)

print("Sorted array:", sorted_arr)
print("Time Complexity of Bubble Sort: O(n²)")
```

```python
def quick_sort(arr):

    if len(arr) <= 1:
        return arr

    pivot = arr[len(arr)//2]

    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]

    return quick_sort(left) + middle + quick_sort(right)


arr = [38, 27, 43, 3, 9, 82, 10]
sorted_arr = quick_sort(arr)

print("Sorted array:", sorted_arr)
print("Time Complexity of Quick Sort: O(n log n)")
```
